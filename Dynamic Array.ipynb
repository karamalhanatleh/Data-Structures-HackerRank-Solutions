```python
#!/bin/python3

def dynamicArray(n, queries):
    # Initialize an array of n empty lists
    arr = [[] for _ in range(n)]
    lastAnswer = 0  # Initialize lastAnswer to 0
    answers = []  # List to store results of type 2 queries
    
    # Loop through each query
    for query in queries:
        q_type, x, y = query  # Extract query type, x, and y from the query
        idx = (x ^ lastAnswer) % n  # Calculate the index using XOR and modulo
        
        # If query type is 1, append y to the selected subarray
        if q_type == 1:
            arr[idx].append(y)
        
        # If query type is 2, calculate lastAnswer and store the result
        elif q_type == 2:
            # Calculate the new lastAnswer from the selected subarray
            lastAnswer = arr[idx][y % len(arr[idx])]
            answers.append(lastAnswer)  # Add lastAnswer to results list
    
    return answers  # Return the list of all lastAnswer values from type 2 queries

if __name__ == '__main__':
    # Take n (number of empty lists) and q (number of queries) as input
    first_multiple_input = input().rstrip().split()
    n = int(first_multiple_input[0])
    q = int(first_multiple_input[1])

    queries = []  # List to store all queries

    # Read each query and add to the queries list
    for _ in range(q):
        queries.append(list(map(int, input().rstrip().split())))

    # Execute the dynamicArray function and get the results
    result = dynamicArray(n, queries)
    
    # Print each result from type 2 queries on a new line
    print('\n'.join(map(str, result)))
```
