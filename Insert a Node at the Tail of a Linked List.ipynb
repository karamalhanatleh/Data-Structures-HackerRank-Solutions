# Insert a Node at the Tail of a Linked List

```python
#!/bin/python3

import math
import os
import random
import re
import sys

class SinglyLinkedListNode:
    def __init__(self, node_data):
        self.data = node_data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

def print_singly_linked_list(node, sep, fptr):
    while node:
        fptr.write(str(node.data))
        node = node.next
        if node:
            fptr.write(sep)

# Complete the insertNodeAtTail function below.
def insertNodeAtTail(head, data):
    # Create a new node with the given data
    new_node = SinglyLinkedListNode(data)
    
    # Check if the head is None (list is empty)
    if head is None:
        return new_node  # The new node becomes the head
    
    # If the list is not empty, traverse to the end
    current_node = head
    while current_node.next:
        current_node = current_node.next  # Move to the next node
    
    # Attach the new node to the end of the list
    current_node.next = new_node
    
    return head  # Return the head of the modified list

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    llist_count = int(input())  # Read the number of elements

    llist = SinglyLinkedList()  # Create a new linked list

    for i in range(llist_count):
        llist_item = int(input())  # Read each value
        llist_head = insertNodeAtTail(llist.head, llist_item)  # Insert at tail
        llist.head = llist_head  # Update the head

    print_singly_linked_list(llist.head, '\n', fptr)  # Print the list
    fptr.write('\n')
    fptr.close()
```
