# 2D Array - DS

```python
def hourglassSum(arr):
    max_sum = -63

    # Iterate over all possible hourglass positions
    for i in range(4):
        for j in range(4):
            # Calculate  hourglass sum
            current_sum = (
                arr[i][j] + arr[i][j+1] + arr[i][j+2] +  # Top row
                arr[i+1][j+1] +                          # Middle
                arr[i+2][j] + arr[i+2][j+1] + arr[i+2][j+2]  # Bottom row
            )
            # Update max_sum
            max_sum = max(max_sum, current_sum)
            
    return max_sum

if __name__ == '__main__':
    arr = []

    # Input for the 6x6 array
    for _ in range(6):
        arr.append(list(map(int, input().rstrip().split())))

    # Get the result from the function
    result = hourglassSum(arr)

    # Output the result directly to the console
    print(result)
```
